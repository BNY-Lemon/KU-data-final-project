```python
master_df = pd.DataFrame()
for i in data_id:
    df = pd.read_table(f'https://wals.info/feature/{i}.tab', header=5)
    url = 'https://wals.info/feature/1A.html#2/19.3/153.1'
    response = requests.get(f'https://wals.info/feature/{i}.html#2/19.3/153.1')
    soup = bs(response.text, 'html.parser')
    title = ''
    titles = soup.find_all('title')
    for i in titles:
        title = i.text.strip()[26:]
    df["title"] = title
    master_df = master_df.append(df)
    print(title)
```
